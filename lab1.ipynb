```python
list1 = [2, 7, 4, 1, 3, 6]
print(list1)

pairs = []
for i in range (len(list1) - 1 ):
  for j in range(i+1,len(list1)):
    if list1[i]+list1[j]==10:
      pairs.append((list1[i],list1[j]))

print(pairs)
```

```python
matrix = int(input("enter order of square matrix "))
mat = [[0] * matrix for _ in range(matrix)]
for i in range((matrix)):
  for j in range((matrix)):
    m = int(input())
    mat[i][j]= m

print(mat)
```

```python
ch = __builtins__.list(input("enter a string "))
print(ch)

count = {}
for c in ch:

    if c in count:
      count[c] = count[c] + 1
    else:
      count[c] = 1

largest = max(count, key=count.get)

print(largest, count[largest])
```

```python
import random
import statistics

numbers = [random.randint(1, 10) for _ in range(25)]

print(numbers)

print(statistics.mean(numbers))
print(statistics.mode(numbers))
print(statistics.variance(numbers))
```
